```python
def get_dataframe(playlist_URI: str) -> pd.core.frame.DataFrame:
    artist_names, artist_pops = [], []
    track_names, track_pops, track_vals, track_dances, track_modes = [], [], [], [], []
    for track in sp.playlist_tracks(playlist_URI)["items"]:
        #Track URI
        track_uri = track["track"]["uri"]

        #Track Name, Popularity, Valence, Danceability, Mode
        track_names.append(track["track"]["name"])
        track_pops.append(track["track"]["popularity"])
        track_vals.append(sp.audio_features(track_uri)[0]["valence"])
        track_dances.append(sp.audio_features(track_uri)[0]["danceability"])
        track_modes.append(sp.audio_features(track_uri)[0]["mode"])

        #Artist URI (Main)
        artist_uri = track["track"]["artists"][0]["uri"]
        artist_info = sp.artist(artist_uri)

        #Artist Name and Popularity
        # artist_name = artist_info["name"]
        artist_names.append(artist_info["name"])
        # artist_pop = artist_info["popularity"]
        artist_pops.append(artist_info["popularity"])

    assert len(artist_names) == len(artist_pops) == len(track_names) == len(track_pops) == len(track_vals) == len(track_dances) == len(track_modes)
    
    #Create Dataframe with parsed lists
    df_data = pd.DataFrame()
    df_data['artist'] = pd.Series(artist_names, dtype=str)
    df_data['artist popularity'] = pd.Series(artist_pops, dtype=int)
    df_data['name'] = pd.Series(track_names, dtype=str)
    df_data['popularity'] = pd.Series(track_pops, dtype=int)
    df_data['valence'] = pd.Series(track_vals, dtype=float)
    df_data['danceability'] = pd.Series(track_dances, dtype=float)
    df_data['modes'] = pd.Series(track_modes, dtype=int)
    
    return df_data
```
